```python
import pandas as pd
```

```python
# Se cargan los datos en un dataframe y mostramos la información del dataset
data = pd.read_csv('dataset.csv')
data.info()
```

```python
# Se muestran los nombres de los atributos
data.columns
```

```python
# Estadísticas básicas de las variables numéricas del dataset
data.describe()
```

```python
# Se muestran las primeras 5 muestras del dataset
data.head()
```

```python
# La primer columna es una serie de índices la cual no nos resulta de utilidad
# Se procede a eliminarla
data.drop(data.columns[0], 1, inplace=True)
data.head()
```

```python
# La variable categórica TGB_measured da información acerca de si existe una medición en la 
# variable TGB, la cual fue eliminada anteriormente en la fase de imputación, por tanto 
# procedemos a eliminar dicha columna también
data.drop('TBG_measured', 1, inplace=True)
data.head()
```

```python
# Basados en la información obtenida del conjunto de datos original, el cual se encuentra en
# https://archive.ics.uci.edu/ml/machine-learning-databases/thyroid-disease/sick-euthyroid.data
# la primer columna de dicho dataset indica la clase correspondiente a la muestra
# Fácilmente se puede observar que la cantidad de muestras de la clase sick-euthyroid es de 293
# mientras que la cantidad de muestras de clase negative es de 2870.
# En el dataset actual no se cuenta con dicha columna, por tanto se procede a agregarla al conjunto
```

```python
# Se crea la lista de clases
sick_euthyroid = ['sick-euthyroid' for i in range(293)]
negative = ['negative' for i in range(2870)]

classes = sick_euthyroid + negative
classes
```

```python
# Se agrega al dataset
data['classes'] = classes
data.head()
```

```python
# Se hace un conteo de muestras por clases
data.classes.value_counts()
```

```python
# Se convierte variable sex a valores de 0 y 1
data['sex'] = data['sex'].map({'F': 0, 'M': 1})
```

```python
# Se convierten variables con valires f y t a valores de 0 y 1

data['sick'] = data['sick'].map({'f': 0, 't': 1})
data['tumor'] = data['tumor'].map({'f': 0, 't': 1})
data['goitre'] = data['goitre'].map({'f': 0, 't': 1})
data['lithium'] = data['lithium'].map({'f': 0, 't': 1})
data['pregnant'] = data['pregnant'].map({'f': 0, 't': 1})
data['on_thyroxine'] = data['on_thyroxine'].map({'f': 0, 't': 1})
data['thyroid_surgery'] = data['thyroid_surgery'].map({'f': 0, 't': 1})
data['query_hypothyroid'] = data['query_hypothyroid'].map({'f': 0, 't': 1})
data['query_hyperthyroid'] = data['query_hyperthyroid'].map({'f': 0, 't': 1})
data['query_on_thyroxine'] = data['query_on_thyroxine'].map({'f': 0, 't': 1})
data['on_antithyroid_medication'] = data['on_antithyroid_medication'].map({'f': 0, 't': 1})
```

```python
# Se convierte variables con valores n e y a valores de 0 y 1
data['T3_measured'] = data['T3_measured'].map({'n': 0, 'y': 1})
data['TSH_measured'] = data['TSH_measured'].map({'n': 0, 'y': 1})
data['TT4_measured'] = data['TT4_measured'].map({'n': 0, 'y': 1})
data['T4U_measured'] = data['T4U_measured'].map({'n': 0, 'y': 1})
data['FTI_measured'] = data['FTI_measured'].map({'n': 0, 'y': 1})
```

```python
# Se convierte variable classes a valores de 0 y 1
data['classes'] = data['classes'].map({'negative': 0, 'sick-euthyroid': 1})
```
